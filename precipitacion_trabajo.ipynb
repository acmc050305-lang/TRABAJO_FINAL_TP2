```python
# percentiles para los 3 años
ANIOS   = [2017, 2019, 2023]
SERIES  = [serie, serie2, serie3]
DFS     = [serie_df, serie_df2, serie_df3]
LABELS  = ["2017 (Niño Costero)", "2019 (Referencia)", "2023 (Niño/Yaku)"]
COLORES = ["#1f77b4", "#2ca02c", "#d62728"]

# Percentiles globales (sobre toda la serie de cada año)
pct = {}
for anio, s in zip(ANIOS, SERIES):
    pct[anio] = {
        "P75": float(s.quantile(0.75)),
        "P90": float(s.quantile(0.90)),
        "P95": float(s.quantile(0.95)),
        "P99": float(s.quantile(0.99)),
    }

resumen_pct = pd.DataFrame(pct).T
resumen_pct.index.name = "Año"
print("Percentiles globales por año")
print(resumen_pct.round(2))
```

## Umbrales

```python
# Tabla de umbrales críticos
print("\nUmbrales críticos de precipitación – Cuenca Río Piura\n")
for anio in ANIOS:
    p = pct[anio]
    print(f"  {anio}")
    print(f"    Moderadamente lluvioso (P75–P90): {p['P75']:.2f} – {p['P90']:.2f} mm")
    print(f"    Lluvioso              (P90–P95): {p['P90']:.2f} – {p['P95']:.2f} mm")
    print(f"    Muy lluvioso          (P95–P99): {p['P95']:.2f} – {p['P99']:.2f} mm")
    print(f"    Extremadamente lluvioso  (>P99): > {p['P99']:.2f} mm")
    print()
```

```python
# Función de alerta aplicada a los 3 años
def asignar_alerta(valor, p90, p95, p99):
    if valor >= p99:
        return "Roja"
    elif valor >= p95:
        return "Naranja"
    elif valor >= p90:
        return "Amarilla"
    else:
        return "Normal"

for anio, df in zip(ANIOS, DFS):
    p = pct[anio]
    df["alerta"] = df["precip"].apply(
        lambda x: asignar_alerta(x, p["P90"], p["P95"], p["P99"])
    )

# Conteo de días por nivel de alerta en cada año
print("Días por nivel de alerta")
niveles = ["Roja", "Naranja", "Amarilla", "Normal"]
tabla_alertas = pd.DataFrame(index=niveles)
for anio, df in zip(ANIOS, DFS):
    tabla_alertas[anio] = [
        (df["alerta"] == n).sum() for n in niveles
    ]
print(tabla_alertas)
```

```python
# Gráfico umbrales de alerta
fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=False)

meses_ticks  = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
meses_labels = ["Ene","Feb","Mar","Abr","May","Jun",
                "Jul","Ago","Sep","Oct","Nov","Dic"]

colores_alerta = {
    "Roja":     "#d62728",
    "Naranja":  "#ff7f0e",
    "Amarilla": "#f5c518",
    "Normal":   "#aec7e8",
}

for idx, (anio, df, label) in enumerate(zip(ANIOS, DFS, LABELS)):
    ax  = axes[idx]
    p   = pct[anio]
    doy = np.arange(1, len(df) + 1)

    # Colorear fondo según nivel de alerta
    for i, (_, row) in enumerate(df.iterrows()):
        ax.axvspan(doy[i] - 0.5, doy[i] + 0.5,
                    color=colores_alerta[row["alerta"]], alpha=0.25, linewidth=0)

    # Serie de precipitación
    ax.plot(doy, df["precip"].values, color="black", linewidth=0.8, zorder=5)

    # Líneas de umbral
    ax.axhline(p["P90"], color="gold",   linestyle="--", linewidth=1.2,
                label=f"P90 = {p['P90']:.1f} mm")
    ax.axhline(p["P95"], color="orange", linestyle="--", linewidth=1.2,
                label=f"P95 = {p['P95']:.1f} mm")
    ax.axhline(p["P99"], color="red",    linestyle="--", linewidth=1.2,
                label=f"P99 = {p['P99']:.1f} mm")

    ax.set_title(label, fontsize=12, fontweight="bold")
    ax.set_ylabel("Precipitación (mm/día)", fontsize=9)
    ax.set_xticks(meses_ticks)
    ax.set_xticklabels(meses_labels)
    ax.grid(alpha=0.3)
    ax.legend(fontsize=8, loc="upper right")

# Leyenda de colores de alerta
parches = [
    mpatches.Patch(color=colores_alerta["Roja"],     alpha=0.5, label="Alerta Roja (>P99)"),
    mpatches.Patch(color=colores_alerta["Naranja"],  alpha=0.5, label="Alerta Naranja (P95–P99)"),
    mpatches.Patch(color=colores_alerta["Amarilla"], alpha=0.5, label="Alerta Amarilla (P90–P95)"),
    mpatches.Patch(color=colores_alerta["Normal"],   alpha=0.5, label="Normal (<P90)"),
]
fig.legend(handles=parches, loc="lower center", ncol=4,
            fontsize=9, bbox_to_anchor=(0.5, -0.02))

fig.suptitle(
    "Umbrales de alerta por precipitación diaria\nCuenca del Río Piura – 2017, 2019 y 2023",
    fontsize=14, fontweight="bold"
)
plt.tight_layout()
plt.show()
```

```python
"""
Descarga y procesamiento de Avisos Meteorológicos del SENAMHI
Cuenca del Río Piura – Años 2017, 2019 y 2023

Uso:
    pip install requests beautifulsoup4 pandas openpyxl
    python descargar_avisos_senamhi.py

Salida:
    avisos_senamhi_piura.xlsx  → todos los avisos de lluvia/precipitación
                                    para Piura en los 3 años
"""

import requests
import pandas as pd
import time
from bs4 import BeautifulSoup

# Configuración

ANIOS = [2017, 2019, 2023]

# Palabras clave para filtrar avisos relevantes a Piura (precipitación)
PALABRAS_LLUVIA = [
    "lluvia", "precipitac", "costa norte", "piura", "tumbes",
    "sierra norte", "lluvias intensas"
]

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )
}

# Funciones

def obtener_avisos_historicos(anio: int) -> list[dict]:
    """
    Descarga la tabla de avisos históricos del SENAMHI para un año dado.
    Retorna una lista de diccionarios con los datos de cada aviso.
    """
    # URL del histórico nacional (los avisos de Piura están aquí con código regional 022)
    url = f"https://www.senamhi.gob.pe/?p=aviso-meteorologico-historico&anio={anio}"

    # Para años anteriores a 2021, la URL del histórico es diferente
    if anio <= 2023:
        url_historico = "https://www.senamhi.gob.pe/?p=aviso-meteorologico-historico"
    
    print(f"\n  Consultando año {anio}...")
    
    try:
        resp = requests.get(url_historico, headers=HEADERS, timeout=30)
        resp.raise_for_status()
    except requests.RequestException as e:
        print(f"  ✗ Error al conectar: {e}")
        return []

    soup = BeautifulSoup(resp.text, "html.parser")
    tabla = soup.find("table")
    if not tabla:
        print(f"  ✗ No se encontró tabla de avisos para {anio}")
        return []

    filas = tabla.find_all("tr")
    avisos = []

    for fila in filas[2:]:  # Saltar cabeceras
        celdas = fila.find_all("td")
        if len(celdas) < 6:
            continue

        descripcion = celdas[0].get_text(strip=True)
        nro          = celdas[1].get_text(strip=True)
        emision      = celdas[2].get_text(strip=True)
        inicio       = celdas[3].get_text(strip=True)
        fin          = celdas[4].get_text(strip=True)
        duracion     = celdas[5].get_text(strip=True) if len(celdas) > 5 else ""
        nivel        = celdas[6].get_text(strip=True) if len(celdas) > 6 else ""

        # Filtrar solo avisos del año solicitado
        if emision and not emision.startswith(str(anio)):
            continue

        # Filtrar solo avisos relacionados con lluvia/precipitación en Piura/Costa Norte
        desc_lower = descripcion.lower()
        if not any(p in desc_lower for p in PALABRAS_LLUVIA):
            continue

        # Obtener link del detalle si existe
        link_tag = celdas[0].find("a")
        link = link_tag["href"] if link_tag and "href" in link_tag.attrs else ""

        avisos.append({
            "año":         anio,
            "nro":         nro,
            "descripcion": descripcion,
            "fecha_emision": emision,
            "fecha_inicio":  inicio,
            "fecha_fin":     fin,
            "duracion":      duracion,
            "nivel":         nivel,
            "link":          f"https://www.senamhi.gob.pe{link}" if link.startswith("?") else link,
        })

    print(f"  ✓ {len(avisos)} avisos de precipitación encontrados en {anio}")
    return avisos


def obtener_avisos_piura_directo(anio: int) -> list[dict]:
    """
    Consulta directamente la página de avisos de Piura
    usando la URL regional dp=piura.
    Funciona mejor para años más recientes (2021+).
    """
    url = f"https://www.senamhi.gob.pe/main.php?dp=piura&p=avisos-meteorologicos&anio={anio}"
    print(f"\n  Consultando avisos Piura {anio} (URL regional)...")

    try:
        resp = requests.get(url, headers=HEADERS, timeout=30)
        resp.raise_for_status()
    except requests.RequestException as e:
        print(f"  ✗ Error: {e}")
        return []

    soup = BeautifulSoup(resp.text, "html.parser")
    tabla = soup.find("table")
    if not tabla:
        print(f"  ✗ No se encontró tabla para {anio} en URL regional")
        return []

    filas = tabla.find_all("tr")
    avisos = []

    for fila in filas[2:]:
        celdas = fila.find_all("td")
        if len(celdas) < 5:
            continue

        descripcion  = celdas[0].get_text(strip=True)
        nro          = celdas[1].get_text(strip=True)
        emision      = celdas[2].get_text(strip=True)
        inicio       = celdas[3].get_text(strip=True)
        fin          = celdas[4].get_text(strip=True)
        duracion     = celdas[5].get_text(strip=True) if len(celdas) > 5 else ""
        nivel        = celdas[6].get_text(strip=True) if len(celdas) > 6 else ""

        if emision and not emision.startswith(str(anio)):
            continue

        desc_lower = descripcion.lower()
        if not any(p in desc_lower for p in PALABRAS_LLUVIA):
            continue

        link_tag = celdas[0].find("a")
        link = link_tag["href"] if link_tag and "href" in link_tag.attrs else ""

        avisos.append({
            "año":           anio,
            "nro":           nro,
            "descripcion":   descripcion,
            "fecha_emision": emision,
            "fecha_inicio":  inicio,
            "fecha_fin":     fin,
            "duracion":      duracion,
            "nivel":         nivel,
            "link":          f"https://www.senamhi.gob.pe{link}" if link.startswith("?") else link,
        })

    print(f"  ✓ {len(avisos)} avisos encontrados")
    return avisos


def expandir_fechas(df: pd.DataFrame) -> pd.DataFrame:
    """
    Expande cada aviso en filas diarias (una fila por fecha cubierta).
    Útil para hacer el merge con los datos CHIRPS por fecha.
    """
    filas_expandidas = []
    for _, row in df.iterrows():
        try:
            inicio = pd.to_datetime(row["fecha_inicio"])
            fin    = pd.to_datetime(row["fecha_fin"])
        except Exception:
            continue
        for fecha in pd.date_range(inicio, fin):
            filas_expandidas.append({
                "fecha":       fecha.date(),
                "año":         row["año"],
                "nro":         row["nro"],
                "descripcion": row["descripcion"],
                "nivel_senamhi": row["nivel"],
                "duracion":    row["duracion"],
            })
    return pd.DataFrame(filas_expandidas)


# ── Script principal ──────────────────────────────────────────────────────────

if __name__ == "__main__":
    print("=" * 60)
    print("  Descarga de Avisos Meteorológicos SENAMHI – Piura")
    print("  Años: 2017, 2019, 2023")
    print("=" * 60)

    todos_avisos = []

    for anio in ANIOS:
        # Intentar primero con URL del histórico nacional (incluye 2017 y 2019)
        avisos = obtener_avisos_historicos(anio)

        # Si no se obtuvieron datos, intentar con URL regional
        if not avisos:
            avisos = obtener_avisos_piura_directo(anio)

        todos_avisos.extend(avisos)
        time.sleep(1.5)  # Pausa entre requests para no saturar el servidor

    if not todos_avisos:
        print("\n⚠ No se obtuvieron avisos automáticamente.")
        print("  Ver instrucciones manuales al final de este script.")
    else:
        df = pd.DataFrame(todos_avisos)

        # ── Tabla de avisos (una fila por aviso) ──
        df.to_excel("avisos_senamhi_piura.xlsx", index=False)
        print(f"\n✓ Guardado: avisos_senamhi_piura.xlsx  ({len(df)} avisos)")

        # ── Tabla expandida por días (para merge con CHIRPS) ──
        df_diario = expandir_fechas(df)
        df_diario.to_excel("avisos_senamhi_piura_diario.xlsx", index=False)
        print(f"✓ Guardado: avisos_senamhi_piura_diario.xlsx  ({len(df_diario)} filas-día)")

        # ── Resumen por año y nivel ──
        print("\n=== Resumen de avisos de precipitación encontrados ===")
        resumen = (
            df.groupby(["año", "nivel"])
            .size()
            .reset_index(name="cantidad")
        )
        print(resumen.to_string(index=False))


# ── INSTRUCCIONES MANUALES (si el scraping falla) ─────────────────────────────
"""
Si el script no obtiene datos automáticamente (la web puede bloquear bots),
sigue estos pasos manuales:

1. Abre en tu navegador:
   - https://www.senamhi.gob.pe/?p=aviso-meteorologico-historico
     (contiene avisos 2017–2020 aproximadamente)
   - https://www.senamhi.gob.pe/main.php?dp=piura&p=avisos-meteorologicos
     (contiene avisos recientes incluyendo 2023)

2. Para cada año, filtra los avisos que contengan:
   "LLUVIA", "PRECIPITACIONES", "COSTA NORTE", "PIURA"

3. Copia los datos a un Excel con estas columnas:
   fecha | descripcion | nivel_senamhi
   
   Donde "fecha" es la fecha de INICIO del aviso (YYYY-MM-DD)
   y "nivel_senamhi" es: AMARILLO, NARANJA o ROJO

4. Guarda el archivo como: avisos_senamhi_piura_diario.xlsx

5. Usa ese archivo en la Celda 8 del notebook (ya está preparada).

FECHAS CLAVE A BUSCAR (eventos extremos documentados):
   - 2017: Enero–Abril (Niño Costero, peores en Febrero–Marzo)
   - 2023: Febrero–Abril (Ciclón Yaku, peores en Marzo)
   - 2019: Febrero–Marzo (año de referencia, menos eventos)

NOTA IMPORTANTE:
   El SENAMHI implementó el sistema de avisos meteorológicos 
   formalmente alrededor de 2017-2018, por lo que el año 2017
   puede tener cobertura limitada o incompleta en el archivo histórico.
   Para ese año, también puedes consultar los informes del INDECI:
   https://portal.indeci.gob.pe/informes-de-emergencia/
"""
```

## Comparación avisos met

```python
# ============================================================
# CELDA 8 – COMPARACIÓN CHIRPS vs AVISOS SENAMHI (DATOS REALES)
# Reemplaza la Celda 8 comentada del código anterior
# ============================================================
# Asume que ya tienes en tu carpeta:
#   avisos_senamhi_piura_diario.xlsx  (890 filas, descargado del SENAMHI)
# Y que ya tienes cargadas:
#   serie, serie2, serie3   → series diarias 2017, 2019, 2023
#   pct                     → diccionario de percentiles por año

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.ticker import MaxNLocator

# ── 1. Cargar avisos SENAMHI ─────────────────────────────────────────────────
RUTA_AVISOS = r"avisos_senamhi_piura_diario.xlsx"   # ajusta ruta si es necesario

avisos = pd.read_excel(RUTA_AVISOS, parse_dates=["fecha"])
avisos["fecha"] = pd.to_datetime(avisos["fecha"]).dt.normalize()

# Estandarizar nombre de columna de nivel (puede llamarse "nivel" o "nivel_senamhi")
col_nivel = [c for c in avisos.columns if "nivel" in c.lower()][0]
avisos = avisos.rename(columns={col_nivel: "nivel_senamhi"})
avisos["nivel_senamhi"] = avisos["nivel_senamhi"].str.upper().str.strip()

print(f"✓ Avisos cargados: {len(avisos)} filas-día")
print(avisos["nivel_senamhi"].value_counts())


# ── 2. Construir series diarias con nivel CHIRPS para los 3 años ─────────────
ANIOS  = [2017, 2019, 2023]
SERIES = [serie, serie2, serie3]

# Mapeo de niveles SENAMHI → categorías equivalentes CHIRPS
# SENAMHI usa: AMARILLO=lluvioso, NARANJA=muy lluvioso, ROJO=extremo
# CHIRPS usa:  Amarilla (P90-P95), Naranja (P95-P99), Roja (>P99)
EQUIV = {"AMARILLO": "Amarilla", "NARANJA": "Naranja", "ROJO": "Roja"}

dfs_chirps = []
for anio, s in zip(ANIOS, SERIES):
    p = pct[anio]
    df_tmp = s.to_dataframe(name="precip").reset_index()
    df_tmp.columns = ["fecha", "precip"]
    df_tmp["fecha"] = pd.to_datetime(df_tmp["fecha"]).dt.normalize()
    df_tmp["año"]   = anio

    def nivel_chirps(v):
        if v >= p["P99"]: return "Roja"
        if v >= p["P95"]: return "Naranja"
        if v >= p["P90"]: return "Amarilla"
        return "Normal"

    df_tmp["nivel_chirps"] = df_tmp["precip"].apply(nivel_chirps)
    dfs_chirps.append(df_tmp)

serie_total = pd.concat(dfs_chirps, ignore_index=True)

# ── 3. Merge CHIRPS con SENAMHI por fecha ────────────────────────────────────
# Solo los días donde SENAMHI emitió aviso
comparacion = serie_total.merge(
    avisos[["fecha", "nivel_senamhi"]].drop_duplicates("fecha"),
    on="fecha",
    how="inner"
)

# Normalizar nivel SENAMHI al mismo esquema que CHIRPS
comparacion["nivel_senamhi_norm"] = comparacion["nivel_senamhi"].map(EQUIV)
comparacion = comparacion.dropna(subset=["nivel_senamhi_norm"])

print(f"\n✓ Días con aviso SENAMHI que coinciden con datos CHIRPS: {len(comparacion)}")
print(comparacion[["fecha", "año", "precip", "nivel_chirps", "nivel_senamhi_norm"]].head(10).to_string(index=False))


# ── 4. Tabla de coincidencias ─────────────────────────────────────────────────
ORDEN = ["Amarilla", "Naranja", "Roja"]

tabla_coinc = pd.crosstab(
    comparacion["nivel_chirps"],
    comparacion["nivel_senamhi_norm"],
    margins=True,
    margins_name="Total"
)
# Reordenar columnas
cols_presentes = [c for c in ORDEN if c in tabla_coinc.columns] + ["Total"]
tabla_coinc = tabla_coinc.reindex(columns=cols_presentes, fill_value=0)
rows_presentes = [r for r in ORDEN + ["Normal"] if r in tabla_coinc.index] + ["Total"]
tabla_coinc = tabla_coinc.reindex(rows_presentes, fill_value=0)

print("\n=== Tabla de coincidencias: CHIRPS (filas) vs SENAMHI (columnas) ===")
print(tabla_coinc)

# Tasa de concordancia
dias_conc = sum(
    tabla_coinc.loc[n, n]
    for n in ORDEN
    if n in tabla_coinc.index and n in tabla_coinc.columns
)
total_dias = tabla_coinc.loc["Total", "Total"]
tasa = 100 * dias_conc / total_dias if total_dias > 0 else 0
print(f"\nCoincidencias exactas de nivel: {dias_conc}/{total_dias} ({tasa:.1f}%)")

# Guardar
tabla_coinc.to_excel("Comparacion_CHIRPS_SENAMHI.xlsx")
print("✓ Guardado: Comparacion_CHIRPS_SENAMHI.xlsx")


# ── 5. Tabla de coincidencias por año ────────────────────────────────────────
print("\n=== Coincidencias por año ===")
resumen_anio = []
for anio in ANIOS:
    sub = comparacion[comparacion["año"] == anio]
    if len(sub) == 0:
        continue
    n_total = len(sub)
    n_coinc = sum(sub["nivel_chirps"] == sub["nivel_senamhi_norm"])
    # "Alineados" = CHIRPS detectó al menos alerta cuando SENAMHI también la emitió
    n_detect = sum(sub["nivel_chirps"] != "Normal")
    resumen_anio.append({
        "Año":                  anio,
        "Días con aviso SENAMHI": n_total,
        "Coincidencia exacta":  n_coinc,
        "% exacta":             round(100 * n_coinc / n_total, 1),
        "CHIRPS detectó evento":n_detect,
        "% detección":          round(100 * n_detect / n_total, 1),
    })

df_resumen_anio = pd.DataFrame(resumen_anio)
print(df_resumen_anio.to_string(index=False))


# ── 6. Gráfico: comparación visual – 3 años juntos ───────────────────────────
COLORES_NIV = {
    "Roja":     "#d62728",
    "Naranja":  "#ff7f0e",
    "Amarilla": "#f5c518",
    "Normal":   "#aec7e8",
}

fig, axes = plt.subplots(3, 1, figsize=(14, 11), sharex=False)

meses_ticks  = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
meses_labels = ["Ene","Feb","Mar","Abr","May","Jun",
                "Jul","Ago","Sep","Oct","Nov","Dic"]
LABELS_ANIO  = ["2017 (Niño Costero)", "2019 (Referencia)", "2023 (Niño/Yaku)"]

for idx, (anio, df_c, label) in enumerate(zip(ANIOS, dfs_chirps, LABELS_ANIO)):
    ax = axes[idx]
    p  = pct[anio]
    doy = np.arange(1, len(df_c) + 1)

    # Fondo por nivel CHIRPS
    for i, row in df_c.iterrows():
        ax.axvspan(doy[i - df_c.index[0]] - 0.5,
                   doy[i - df_c.index[0]] + 0.5,
                   color=COLORES_NIV[row["nivel_chirps"]], alpha=0.22, linewidth=0)

    # Serie de precipitación
    ax.plot(doy, df_c["precip"].values, color="black", linewidth=0.8, zorder=5)

    # Líneas de umbral
    ax.axhline(p["P90"], color="gold",      linestyle="--", linewidth=1.1, label=f"P90={p['P90']:.1f}")
    ax.axhline(p["P95"], color="darkorange", linestyle="--", linewidth=1.1, label=f"P95={p['P95']:.1f}")
    ax.axhline(p["P99"], color="red",        linestyle="--", linewidth=1.1, label=f"P99={p['P99']:.1f}")

    # Marcar días con aviso SENAMHI como triángulos en el eje superior
    sub_av = comparacion[comparacion["año"] == anio].copy()
    sub_av["doy"] = sub_av["fecha"].dt.dayofyear
    y_marker = ax.get_ylim()[1] if ax.get_ylim()[1] > 0 else 5

    for _, row_av in sub_av.iterrows():
        color_av = COLORES_NIV.get(row_av["nivel_senamhi_norm"], "gray")
        ax.plot(row_av["doy"], row_av["precip"] + 0.5,
                marker="v", color=color_av, markersize=5,
                markeredgecolor="black", markeredgewidth=0.4, zorder=10)

    ax.set_title(label, fontsize=12, fontweight="bold")
    ax.set_ylabel("Precipitación (mm/día)", fontsize=9)
    ax.set_xticks(meses_ticks)
    ax.set_xticklabels(meses_labels)
    ax.legend(fontsize=8, loc="upper right")
    ax.grid(alpha=0.25)

# Leyenda general
parches_nivel = [
    mpatches.Patch(color=COLORES_NIV["Roja"],     alpha=0.6, label="Roja  (CHIRPS >P99)"),
    mpatches.Patch(color=COLORES_NIV["Naranja"],  alpha=0.6, label="Naranja (CHIRPS P95–P99)"),
    mpatches.Patch(color=COLORES_NIV["Amarilla"], alpha=0.6, label="Amarilla (CHIRPS P90–P95)"),
    mpatches.Patch(color=COLORES_NIV["Normal"],   alpha=0.6, label="Normal  (CHIRPS <P90)"),
]
parche_senamhi = [
    plt.Line2D([0],[0], marker="v", color="w", markerfacecolor=COLORES_NIV["Roja"],
               markersize=8, markeredgecolor="black", label="▼ Aviso SENAMHI Rojo"),
    plt.Line2D([0],[0], marker="v", color="w", markerfacecolor=COLORES_NIV["Naranja"],
               markersize=8, markeredgecolor="black", label="▼ Aviso SENAMHI Naranja"),
    plt.Line2D([0],[0], marker="v", color="w", markerfacecolor=COLORES_NIV["Amarilla"],
               markersize=8, markeredgecolor="black", label="▼ Aviso SENAMHI Amarillo"),
]

fig.legend(handles=parches_nivel + parche_senamhi,
           loc="lower center", ncol=4, fontsize=8,
           bbox_to_anchor=(0.5, -0.03))

fig.suptitle(
    "Comparación CHIRPS vs Avisos SENAMHI – Cuenca del Río Piura\n"
    "Fondo: nivel detectado por CHIRPS  |  Triángulos ▼: aviso emitido por SENAMHI",
    fontsize=13, fontweight="bold"
)
plt.tight_layout()
plt.show()


# ── 7. Gráfico de barras: coincidencia por año ───────────────────────────────
fig, ax = plt.subplots(figsize=(9, 5))

x     = np.arange(len(ANIOS))
ancho = 0.35

total_av = df_resumen_anio["Días con aviso SENAMHI"].values
coinc_ex = df_resumen_anio["Coincidencia exacta"].values
detectado= df_resumen_anio["CHIRPS detectó evento"].values

b1 = ax.bar(x - ancho/2, total_av, ancho,
            label="Días con aviso SENAMHI", color="#7fbfff", edgecolor="black")
b2 = ax.bar(x + ancho/2, detectado, ancho,
            label="CHIRPS detectó alerta", color="#ff9f7f", edgecolor="black")

# Etiquetas de porcentaje
for i, (tot, det, coinc) in enumerate(zip(total_av, detectado, coinc_ex)):
    ax.text(x[i] - ancho/2, tot + 0.5,
            f"{tot}", ha="center", va="bottom", fontsize=9)
    ax.text(x[i] + ancho/2, det + 0.5,
            f"{det}\n({df_resumen_anio['% detección'].iloc[i]:.0f}%)",
            ha="center", va="bottom", fontsize=8)

ax.set_xticks(x)
ax.set_xticklabels([str(a) for a in ANIOS], fontsize=11)
ax.set_ylabel("Número de días", fontsize=10)
ax.set_title(
    "Días con aviso SENAMHI vs días detectados por CHIRPS\nCuenca del Río Piura",
    fontsize=12, fontweight="bold"
)
ax.legend(fontsize=9)
ax.grid(axis="y", alpha=0.35)
plt.tight_layout()
plt.show()


# ── 8. Heatmap de la tabla de coincidencias ──────────────────────────────────
import matplotlib.colors as mcolors

tabla_plot = tabla_coinc.drop(index="Total", columns="Total", errors="ignore")
tabla_np   = tabla_plot.values.astype(float)

fig, ax = plt.subplots(figsize=(7, 5))
im = ax.imshow(tabla_np, cmap="YlOrRd", aspect="auto")

ax.set_xticks(range(len(tabla_plot.columns)))
ax.set_xticklabels(tabla_plot.columns, fontsize=11)
ax.set_yticks(range(len(tabla_plot.index)))
ax.set_yticklabels(tabla_plot.index, fontsize=11)
ax.set_xlabel("Nivel SENAMHI", fontsize=11)
ax.set_ylabel("Nivel CHIRPS",  fontsize=11)

for i in range(tabla_np.shape[0]):
    for j in range(tabla_np.shape[1]):
        val = int(tabla_np[i, j])
        color_txt = "white" if tabla_np[i, j] > tabla_np.max() * 0.6 else "black"
        ax.text(j, i, str(val), ha="center", va="center",
                fontsize=12, fontweight="bold", color=color_txt)

fig.colorbar(im, ax=ax, label="Número de días")
ax.set_title(
    "Coincidencias entre niveles CHIRPS y SENAMHI\n(días con aviso emitido)",
    fontsize=12, fontweight="bold"
)
plt.tight_layout()
plt.show()

print("\n✓ Análisis de comparación CHIRPS vs SENAMHI completado.")
```

```python
import rioxarray as rxr
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable

# ── Rutas ─────────────────────────────────────────────────────────────────────
path_dem = r"C:\Users\andre\Documents\PROYECTO\DEM\DEM.tif"
path_shp = r"C:\Users\andre\Documents\PROYECTO\datos\cuenca_piura.shp"

# ── 1. Shapefile ───────────────────────────────────────────────────────────────
shp = gpd.read_file(path_shp)

# ── 2. Abrir DEM ───────────────────────────────────────────────────────────────
dem_rx = rxr.open_rasterio(path_dem, masked=True)

# Reproyectar shapefile al CRS del DEM
shp = shp.to_crs(dem_rx.rio.crs)
bounds = shp.total_bounds  # bounds YA en el CRS del DEM

# ── 3. Clip por bbox primero (rápido) ─────────────────────────────────────────
margin = 5000  # metros (no grados, porque el DEM está en UTM)
dem_bbox = dem_rx.rio.clip_box(
    minx=bounds[0] - margin,
    miny=bounds[1] - margin,
    maxx=bounds[2] + margin,
    maxy=bounds[3] + margin,
)
print(f"DEM tras bbox: {dem_bbox.shape}")

# ── 4. Clip exacto con polígono ───────────────────────────────────────────────
dem_clip = dem_bbox.rio.clip(shp.geometry, shp.crs, drop=True)
print(f"DEM recortado: {dem_clip.shape}")

# ── 5. Guardar ────────────────────────────────────────────────────────────────
ruta_salida = r"C:\Users\andre\Documents\PROYECTO\datos\DEM_cuenca_del_rio_Piura.tif"
dem_clip.rio.to_raster(ruta_salida, dtype="float32")
print("✅ DEM recortado guardado")

# ── 6. Preparar arrays ────────────────────────────────────────────────────────
arr_bbox = dem_bbox.values.squeeze().astype(float)
arr_clip = dem_clip.values.squeeze().astype(float)
arr_bbox[arr_bbox <= -9999] = np.nan
arr_clip[arr_clip <= -9999] = np.nan

lons_bbox = dem_bbox.x.values; lats_bbox = dem_bbox.y.values
lons_clip = dem_clip.x.values; lats_clip = dem_clip.y.values

# ── 7. Visualizar con colorbar pegada al mapa ─────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# — Izquierda —
im0 = axes[0].pcolormesh(lons_bbox, lats_bbox, arr_bbox,
                          cmap="terrain", shading="auto",
                          vmin=np.nanmin(arr_bbox), vmax=np.nanmax(arr_bbox))
shp.boundary.plot(ax=axes[0], edgecolor="red", linewidth=1.5)
axes[0].set_title("DEM área de estudio", fontweight="bold")
axes[0].set_xlabel("Longitud (m)"); axes[0].set_ylabel("Latitud (m)")
axes[0].set_aspect("equal")  # ← proporciones correctas en UTM

# Colorbar pegada al mapa
div0 = make_axes_locatable(axes[0])
cax0 = div0.append_axes("right", size="4%", pad=0.08)
plt.colorbar(im0, cax=cax0, label="Altitud (m)")

# — Derecha —
im1 = axes[1].pcolormesh(lons_clip, lats_clip, arr_clip,
                          cmap="terrain", shading="auto",
                          vmin=np.nanmin(arr_clip), vmax=np.nanmax(arr_clip))
shp.boundary.plot(ax=axes[1], edgecolor="red", linewidth=1.5)
axes[1].set_title("DEM recortado – Cuenca río Piura", fontweight="bold")
axes[1].set_xlabel("Longitud (m)"); axes[1].set_ylabel("Latitud (m)")
axes[1].set_aspect("equal")  # ← misma proporción

# Colorbar pegada al mapa
div1 = make_axes_locatable(axes[1])
cax1 = div1.append_axes("right", size="4%", pad=0.08)
plt.colorbar(im1, cax=cax1, label="Altitud (m)")

fig.suptitle("Modelo Digital de Elevación – Cuenca del río Piura",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
import rioxarray as rxr
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.ticker import FuncFormatter

# ── 1. Cargar DEM ya recortado a la cuenca ───────────────────────────────────
dem = rxr.open_rasterio(
    r"C:\Users\andre\Documents\PROYECTO\datos\DEM_cuenca_del_rio_Piura.tif",
    masked=True
).squeeze()  # elimina dimensión de banda

# Reproyectar a EPSG:4326 si no lo está
dem = dem.rio.reproject("EPSG:4326")

# ── 2. Definir umbrales altitudinales ────────────────────────────────────────
LIMITES = {"Baja": (0, 500), "Media": (500, 1500), "Alta": (1500, 9999)}
COLORES_ZONA = {"Alta": "#d73027", "Media": "#fee08b", "Baja": "#1a9850"}

# ── 3. Crear máscara por zona usando el DEM ──────────────────────────────────
mascaras = {}
for zona, (z_min, z_max) in LIMITES.items():
    mask = (dem >= z_min) & (dem < z_max)
    mascaras[zona] = mask
    n_px = int(mask.sum())
    print(f"Zona {zona}: {n_px} píxeles ({z_min}–{z_max} m)")

# ── 4. Función: extraer serie temporal promedio por zona ─────────────────────
def serie_por_zona(pp_ds, dem_reproj, mask_zona):
    """
    Para cada timestep, promedia solo los píxeles dentro de la máscara.
    Interpola DEM a la grilla CHIRPS (0.05°) antes de enmascarar.
    """
    # Interpolar DEM a la resolución de CHIRPS
    dem_interp = dem_reproj.interp(
        x=pp_ds.longitude,
        y=pp_ds.latitude,
        method="nearest"
    )
    mask_interp = (dem_interp >= mask_zona[0]) & (dem_interp < mask_zona[1])

    pp_masked = pp_ds["precip"].where(mask_interp)
    serie = pp_masked.mean(dim=["latitude", "longitude"])
    return serie

# ── 5. Calcular series por zona para cada año ─────────────────────────────────
ANIOS_DS = {2017: pp2017, 2019: pp2019, 2023: pp2023}

series_zonas = {}  # {anio: {"Alta": serie, "Media": serie, "Baja": serie}}

for anio, ds in ANIOS_DS.items():
    series_zonas[anio] = {}
    for zona, lims in LIMITES.items():
        dem_interp = dem.interp(
            x=ds.longitude,
            y=ds.latitude,
            method="nearest"
        )
        mask = (dem_interp >= lims[0]) & (dem_interp < lims[1])
        serie_z = ds["precip"].where(mask).mean(dim=["latitude", "longitude"])
        series_zonas[anio][zona] = serie_z
        print(f"  {anio} – Zona {zona}: media anual = {float(serie_z.mean()):.2f} mm/día")

# ── 6. Gráfico: series temporales por zona y año ─────────────────────────────
meses_ticks  = [1,32,60,91,121,152,182,213,244,274,305,335]
meses_labels = ["Ene","Feb","Mar","Abr","May","Jun",
                "Jul","Ago","Sep","Oct","Nov","Dic"]

fig, axes = plt.subplots(3, 1, figsize=(14, 11), sharex=False)
LABELS_ANIO = ["2017 (Niño Costero)", "2019 (Referencia)", "2023 (Niño/Yaku)"]

for idx, (anio, label) in enumerate(zip([2017, 2019, 2023], LABELS_ANIO)):
    ax = axes[idx]
    doy = np.arange(1, 366)

    for zona in ["Alta", "Media", "Baja"]:
        s = series_zonas[anio][zona].to_pandas()
        ax.plot(np.arange(1, len(s)+1), s.values,
                label=f"Zona {zona}",
                color=COLORES_ZONA[zona],
                linewidth=1.1)

    ax.set_title(label, fontsize=12, fontweight="bold")
    ax.set_ylabel("Precipitación (mm/día)", fontsize=9)
    ax.set_xticks(meses_ticks)
    ax.set_xticklabels(meses_labels)
    ax.legend(fontsize=9, loc="upper right")
    ax.grid(alpha=0.3)

fig.suptitle(
    "Precipitación media diaria por zona altitudinal\nCuenca del río Piura",
    fontsize=14, fontweight="bold"
)
plt.tight_layout()
plt.show()

# ── 7. Tabla de percentiles por zona y año ───────────────────────────────────
print("\nPERCENTILES POR ZONA ALTITUDINAL")
print(f"{'Año':<6} {'Zona':<8} {'P75':>8} {'P90':>8} {'P95':>8} {'P99':>8} {'Máx':>8}")
print("-" * 58)

resumen_zonas = []
for anio in [2017, 2019, 2023]:
    for zona in ["Alta", "Media", "Baja"]:
        s = series_zonas[anio][zona].to_pandas().dropna()
        resumen_zonas.append({
            "Año":   anio,
            "Zona":  zona,
            "P75":   round(float(s.quantile(0.75)), 2),
            "P90":   round(float(s.quantile(0.90)), 2),
            "P95":   round(float(s.quantile(0.95)), 2),
            "P99":   round(float(s.quantile(0.99)), 2),
            "Máx":   round(float(s.max()), 2),
        })
        print(f"{anio:<6} {zona:<8} {s.quantile(0.75):>8.2f} "
              f"{s.quantile(0.90):>8.2f} {s.quantile(0.95):>8.2f} "
              f"{s.quantile(0.99):>8.2f} {s.max():>8.2f}")

df_zonas = pd.DataFrame(resumen_zonas)
df_zonas.to_excel("Percentiles_por_zona_Piura.xlsx", index=False)
print("\nGuardado en Percentiles_por_zona_Piura.xlsx")

# ── 8. Mapa de zonas altitudinales ───────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 7))

# Colorear el DEM por zona
dem_vals = dem.values.squeeze()
zona_mapa = np.full(dem_vals.shape, np.nan)
zona_mapa[(dem_vals >= 0)    & (dem_vals < 500)]  = 1   # Baja
zona_mapa[(dem_vals >= 500)  & (dem_vals < 1500)] = 2   # Media
zona_mapa[(dem_vals >= 1500)]                      = 3   # Alta

from matplotlib.colors import ListedColormap, BoundaryNorm
cmap_zonas = ListedColormap(["#1a9850", "#fee08b", "#d73027"])
norm_zonas = BoundaryNorm([0.5, 1.5, 2.5, 3.5], cmap_zonas.N)

lons = dem.x.values
lats = dem.y.values
im = ax.pcolormesh(lons, lats, zona_mapa,
                   cmap=cmap_zonas, norm=norm_zonas, alpha=0.8)

cuenca.boundary.plot(ax=ax, edgecolor="black", linewidth=1.8)

# Leyenda manual
parches = [
    mpatches.Patch(color="#1a9850", label="Zona Baja  (0–500 m)"),
    mpatches.Patch(color="#fee08b", label="Zona Media (500–1500 m)"),
    mpatches.Patch(color="#d73027", label="Zona Alta  (>1500 m)"),
]
ax.legend(handles=parches, loc="lower left", fontsize=10)

def lon_fmt(x, pos): return f"{abs(x):.1f}°O"
def lat_fmt(y, pos): return f"{abs(y):.1f}°S"
ax.xaxis.set_major_formatter(FuncFormatter(lon_fmt))
ax.yaxis.set_major_formatter(FuncFormatter(lat_fmt))
ax.set_xlabel("Longitud"); ax.set_ylabel("Latitud")
ax.set_title("Zonificación altitudinal – Cuenca del río Piura",
             fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()
```
