```python
pp2017 = xr.open_dataset(r"C:\Users\andre\Documents\PROYECTO\datos\pp_Piura_2017.nc")
pp2019 = xr.open_dataset(r"C:\Users\andre\Documents\PROYECTO\datos\pp_Piura_2019.nc")
pp2023 = xr.open_dataset(r"C:\Users\andre\Documents\PROYECTO\datos\pp_Piura_2023.nc")
print(pp2017)
```

```python
print(float(pp2017.longitude.min()))
print(float(pp2017.longitude.max()))

print(float(pp2017.latitude.min()))
print(float(pp2017.latitude.max()))
```

```python
serie = pp2017["precip"].mean(dim=["latitude","longitude"])
serie2 = pp2019["precip"].mean(dim=["latitude","longitude"])
serie3 = pp2023["precip"].mean(dim=["latitude","longitude"])
print(serie)
print(serie2)
print(serie3)
```

```python
serie_df = serie.to_dataframe()
serie_df2 = serie2.to_dataframe()
serie_df3 = serie3.to_dataframe()
print(serie_df.head())
```

```python
s2017 = serie.to_pandas().reset_index(drop=True)
s2019 = serie2.to_pandas().reset_index(drop=True)
s2023 = serie3.to_pandas().reset_index(drop=True)

dias = range(1, len(s2017)+1)

plt.figure(figsize=(14,5))

plt.plot(dias, s2017, label="2017 (Niño Costero)")
plt.plot(dias, s2019, label="2019 (Referencia)")
plt.plot(dias, s2023, label="2023 (Niño)")

plt.xlabel("Día del año")
plt.ylabel("Precipitación media (mm/día)")
plt.title("Comparación de precipitación diaria en la cuenca del río Piura")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

```python
fig, ax = plt.subplots(
    3, 1,
    figsize=(12,8),
    sharex=True,
    sharey=True
)

# 2017
ax[0].plot(serie.time.dt.dayofyear, serie)
ax[0].set_title("2017 (Niño Costero)")
ax[0].set_ylabel("mm/día")
ax[0].grid(alpha=0.3)

# 2019
ax[1].plot(serie2.time.dt.dayofyear, serie2)
ax[1].set_title("2019 (Referencia)")
ax[1].set_ylabel("mm/día")
ax[1].grid(alpha=0.3)

# 2023
ax[2].plot(serie3.time.dt.dayofyear, serie3)
ax[2].set_title("2023 (Niño)")
ax[2].set_ylabel("mm/día")
ax[2].grid(alpha=0.3)

# Meses
meses = [1,32,60,91,121,152,182,213,244,274,305,335]
nombres = ['Ene','Feb','Mar','Abr','May','Jun',
            'Jul','Ago','Sep','Oct','Nov','Dic']

for a in ax:
    a.set_xticks(meses)
    a.set_xticklabels(nombres)

ax[2].set_xlabel("Mes")

plt.suptitle(
    "Precipitación media diaria en la cuenca del río Piura",
    fontsize=16
)

plt.tight_layout()
plt.show()
```

```python
# Precipitación promedio temporal
piura_mean = pp2017['precip'].mean(dim='time')
piura_mean2 = pp2019['precip'].mean(dim='time')
piura_mean3 = pp2023['precip'].mean(dim='time')

# Graficar
plt.figure(figsize=(8,6))

piura_mean.plot(cmap='Blues')

plt.title("Precipitación media - Cuenca del río Piura")
plt.show()
```

```python
vmin = min(
    float(piura_mean.min()),
    float(piura_mean2.min()),
    float(piura_mean3.min())
)

vmax = max(
    float(piura_mean.max()),
    float(piura_mean2.max()),
    float(piura_mean3.max())
)

print(vmin, vmax)
```

```python
# Formato coordenadas
def lon_formatter(x, pos):
    return f"{abs(x):.1f}°O"

def lat_formatter(y, pos):
    return f"{abs(y):.1f}°S"

# Figura
fig, ax = plt.subplots(3,1,
    figsize=(10, 15),
    sharex=True,
    sharey=True)

# Mapa 2017
piura_mean.plot(
    ax=ax[0],
    cmap="Blues",
    vmin=vmin,
    vmax=vmax,
    add_colorbar=False)

ax[0].set_title(
    "2017 (Niño Costero)",
    fontsize=14,
    fontweight="bold")

# Mapa 2019
piura_mean2.plot(
    ax=ax[1],
    cmap="Blues",
    vmin=vmin,
    vmax=vmax,
    add_colorbar=False)

ax[1].set_title("2019 (Año de referencia)",fontsize=14,fontweight="bold")

# Mapa 2023
im = piura_mean3.plot(
    ax=ax[2],
    cmap="Blues",
    vmin=vmin,
    vmax=vmax,
    add_colorbar=False)

ax[2].set_title("2023 (El Niño)",fontsize=14,fontweight="bold")

# Contorno de la cuenca
for a in ax:
    cuenca.boundary.plot(ax=a,edgecolor="red",linewidth=1.5)
    a.xaxis.set_major_formatter(FuncFormatter(lon_formatter))
    a.yaxis.set_major_formatter(FuncFormatter(lat_formatter))
    a.set_xlabel("Longitud")
    a.set_ylabel("Latitud")

# Barra de color
fig.subplots_adjust(
    right=0.85,
    top=0.92,
    hspace=0.18)

cax = fig.add_axes([0.88, 0.15, 0.03, 0.7])
cbar = fig.colorbar(im,cax=cax)
cbar.set_label("Precipitación media diaria (mm/día)",fontsize=11)

# Título
fig.suptitle(
    "Distribución espacial de la precipitación media diaria\n"
    "en la cuenca del río Piura",
    fontsize=18,
    fontweight="bold")

plt.show()
```

```python
p75 = serie.quantile(0.75).values
p90 = serie.quantile(0.90).values
p95 = serie.quantile(0.95).values
p99 = serie.quantile(0.99).values
```

```python
percentiles = pd.DataFrame({
    "Percentil": ["P75", "P90", "P95", "P99"],
    "Valor (mm/día)": [
        p75,
        p90,
        p95,
        p99
    ]
})

print(percentiles)
#percentiles.to_csv(
#    "percentiles_piura_2017.csv",
#    index=False)
```

```python
eventos_extremos = serie_df[
    serie_df["precip"] > p99
]

print(eventos_extremos)
eventos_extremos2 = serie_df[
    serie_df["precip"] > p95
]
print(eventos_extremos2)
```

```python
def alerta(valor, p90, p95, p99):

    if valor >= p99:
        return "Roja"

    elif valor >= p95:
        return "Naranja"

    elif valor >= p90:
        return "Amarilla"

    else:
        return "Normal"

serie_df["alerta"] = serie_df["precip"].apply(
    lambda x: alerta(
        x,
        p90,
        p95,
        p99
    )
)

print(serie_df.head())
```

```python
plt.figure(figsize=(14,5))

plt.plot(
    serie_df.index,
    serie_df["precip"],
    color="black"
)

plt.axhline(
    p90,
    color="gold",
    linestyle="--",
    label="P90"
)

plt.axhline(
    p95,
    color="orange",
    linestyle="--",
    label="P95"
)

plt.axhline(
    p99,
    color="red",
    linestyle="--",
    label="P99"
)

plt.legend()

plt.ylabel("Precipitación (mm/día)")
plt.xlabel("Fecha")

plt.title(
    "Umbrales de alerta para precipitación diaria - Cuenca del río Piura"
)

plt.show()
```
